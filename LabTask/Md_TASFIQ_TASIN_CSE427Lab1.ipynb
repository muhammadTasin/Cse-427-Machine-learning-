```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = np.random.randint(1, 101, 20)
b = np.random.randint(1, 101, 20)

print("Array a:")
print(a)

print("Array b:")
print(b)
```

```python
a = a.reshape(5, 4)
b = b.reshape(5, 4)

print("Matrix a:")
print(a)

print("Matrix b:")
print(b)
```

```python
a_scalar = a * 2
b_scalar = b * 3

print("Scalar multiplication of a:")
print(a_scalar)

print("Scalar multiplication of b:")
print(b_scalar)
```

```python
b_modified = b.reshape(4, 5)

c = np.matmul(a, b_modified)

print("Modified matrix b:")
print(b_modified)

print("Matrix c = a x b_modified:")
print(c)
```

```python
print("Column 2 and 3 of matrix c:")
print(c[:, 1:3])
```

```python
largest = np.max(c)
index = np.unravel_index(np.argmax(c), c.shape)

print("Largest element:", largest)
print("Index of largest element:", index)
```

```python
c_1d = c.flatten()

print("Original matrix c:")
print(c)

print("One dimensional array:")
print(c_1d)
```

```python
temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]

plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.xlabel("Temperature Dhaka")
plt.ylabel("Humidity Dhaka")
plt.title("Temperature vs Humidity")
plt.show()
```

```python
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

plt.figure(figsize=(12, 8))
plt.plot(study_hours, marks)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

```python
temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]

study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
study_marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
subject_marks = [89, 90, 45, 78, 99]

plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.title("Scatter Plot")
plt.xlabel("Temperature")
plt.ylabel("Humidity")

plt.subplot(2, 2, 2)
plt.plot(study_hours, study_marks)
plt.title("Line Plot")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.subplot(2, 2, 3)
plt.barh(subjects, subject_marks)
plt.title("Horizontal Bar Graph")
plt.xlabel("Marks")
plt.ylabel("Subjects")

plt.subplot(2, 2, 4)
plt.bar(subjects, subject_marks, color=['red', 'green', 'blue', 'orange', 'purple'])
plt.title("Vertical Bar Graph")
plt.xlabel("Subjects")
plt.ylabel("Marks")

plt.tight_layout()
plt.show()
```
